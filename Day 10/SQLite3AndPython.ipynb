### SQL And SQLite
#### SQL (Structured Query Language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems. In this lesson, we will cover the basics of SQL and SQLite, including creating databases, tables, and performing various SQL operations.

```python
import sqlite3
```

```python
## Connect to an SQLite database
connection = sqlite3.connect('example.db')
connection
```

```python
## Creating a cursor
# A cursor acts as a control structure or "pointer" that allows you to execute SQL queries and fetch the resulting rows one by one from the database.

cursor = connection.cursor()
```

```python
## Creating a Table (Using same cursor object)

cursor.execute('''
Create Table If Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text
    )
''')

## Commit the changes
connection.commit()
```

```python
## Executing any query example
cursor.execute('''
(SQL Query)
''')
```

```python
## Insert the data in sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
                values('Sher',20,'Data Scientist')
''')

cursor.execute('''
Insert Into employees(name,age,department)
                values('John',25,'Data Engineer')
''')

cursor.execute('''
Insert Into employees(name,age,department)
                values('Jane',30,'Machine Learning')
''')

## Commit the changes
connection.commit()
```

```python
## Query the data from the table
cursor.execute('Select * from employees')
rows = cursor.fetchall() # Get all the records(rows)

## Print the queried data

for row in rows:
    print(row)
```

```python
## Update thedata in the table

cursor.execute('''
Update employees
Set age=27
where name="Sher"
''')

connection.commit()
```

```python
cursor.execute('Select * from employees')
rows = cursor.fetchall() # Get all the records(rows)

## Print the queried data

for row in rows:
    print(row)
```

```python
## Delete the data from the table
cursor.execute('''
Delete from employees
        where name="John"
''')

connection.commit()
```

```python
cursor.execute('Select * from employees')
rows = cursor.fetchall() # Get all the records(rows)

## Print the queried data

for row in rows:
    print(row)
```
