```python
# Pandas-DataFrame And Series
# Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).
```

```python
import pandas as pd
```

```python
## Series
# A Pandas Series is a one-dimentional array-like object that can hold any data type. It is similar to a column in a table

import pandas as pd
data = [1,2,3,4,5]
series=pd.Series(data)
print("Series \n", series)
print(type(series))
# First are the default indexes for the data
```

```python
## Create a series from dictionary
data = {'a':1, 'b': 2, 'c': 3}
series_dict=pd.Series(data)
print(series_dict)
# Keys will be made as index and values will be series values
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data,index=index)
```

```python
## Datafram
# A Pandas DataFrame is a two-dimensional, size-mutable, and tabular data structure with labeled axes (rows and columns).

# CSV & Text: pd.read_csv() for comma-separated values or general delimited text files.

# Excel: pd.read_excel() for .xlsx and .xls files (supports multiple sheets).

# JSON: pd.read_json() for web-based data structures.

# SQL Databases: pd.read_sql() to pull data directly from databases like PostgreSQL, MySQL, or SQLite.

## create a Dataframe from a dictionary of list

data = {   
    'Name': ['Sher', 'Hassan', 'Khan'],
    'Age': [18,19,20],
    'City': ['Quetta', 'Islamabad', 'Zhob']
} # This is basically rows and columns(2d array)

df = pd.DataFrame(data) # This is like data in Excel Sheet
print(df)
print(type(df)) 
```

```python
## Create a Data frame From a list of Dictionaries

data = [
    {'Name': 'Sher','Age': 20, 'City': 'Quetta'},
    {'Name': 'Hassan','Age': 20, 'City': 'Islamabad'},
    {'Name': 'Khan','Age': 20, 'City': 'Zhob'},
    {'Name': 'Ali','Age': 20, 'City': 'Karachi'}
]
df = pd.DataFrame(data)
print(df)
print(type(df))
```
