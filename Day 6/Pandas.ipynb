```python
# Pandas-DataFrame And Series
# Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).
```

```python
## Series
# A Pandas Series is a one-dimentional array-like object that can hold any data type. It is similar to a column in a table

import pandas as pd
data = [1,2,3,4,5]
series=pd.Series(data)
print("Series \n", series)
print(type(series))
# First are the default indexes for the data
```

```python
## Create a series from dictionary
data = {'a':1, 'b': 2, 'c': 3}
series_dict=pd.Series(data)
print(series_dict)
# Keys will be made as index and values will be series values
```

```python
# Directly calling columns
# It will be a pandas series
df['Name']
# type(df['Name'])   #will be series
```

```python
# picking row with index using loc
df.loc[0]
```

```python
# picking column with index using loc
df.iloc[0]
```

```python
# Picking specific using loc (using indexing)
df.loc[0, 'Name']
# usually will work, but it is generally considered bad practice and can lead to unpredictable errors
# df.loc[0, 'Name']: You are telling Pandas exactly which row label (0) and which column label ('Name') to look at in a single operation. It is fast and explicit.
# for indexing like [0, 0] iloc will be better
```

```python
# Picking specific using iloc (using indexing)
df.iloc[0, 2]
```

```python
## Accessing a specified element (.at)
df.at[1, 'Age']  # Row index, column name (R,C)
```

```python
## Accessing a specified element (.iat)
df.iat[1,2]
```

```python
# basically loc and at uses [row index, column name] format and iloc and iat uses [row index, column index]
```
