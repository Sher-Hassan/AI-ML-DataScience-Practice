```python
## Data Serialization
# for conversion
# to work with JSON (usiually used in APIs)
import json
data = {'name': 'Sher', 'age': 25}
json_str=json.dumps(data) # JSON to string
print(json_str)
print(type(json_str))

parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## CSV
# for working with CSV files
import csv

with open('example.csv', mode='w', newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['Sher', '20'])

with open('example.csv', mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
## Regular expression
# re is a standard library module that lets you search, match, split, and replace text using patterns (called regular expressions or regex).
import re

pattern = r'\d+'
text = 'There are 212 apples'
match = re.search(pattern,text)
print(match.group())
```

```python
#Checking if a Path is a file or Directory
import os
path = 'Notes.txt'

if os.path.isfile(path):
    print(f"'{path}' is a file")
elif os.path.isdir(path):
    print(f"'{path}' is a directory")
else:
    print(f"'{path}' is a neither a directory nor a file")
```

```python
## Getting the absolute path
# better than joining (full_final_path = os.path.join(os.getcwd(),dir_name,file_name))
realative_path = 'Notes.txt'
absolute_path = os.path.abspath(realative_path)
print(absolute_path)
```
