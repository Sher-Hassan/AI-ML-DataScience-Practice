### Data Manipulation and Analysis with Pandas
Data manipulation and analysis are key tasks in any data science or data analysis project. Pandas provides a wide range of functions for data manipulation and analysis, making it easier to clean, transform, and extract insights from data. In this lesson, we will cover various data manipulation and analysis techniques using Pandas.

```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
# First five entries
df.head(5)
```

```python
# last 5 entries
df.tail(5)
```

```python
# statistical data
df.describe()
```

```python
# Data types of data
df.dtypes
```

```python
## Handling missing values
# mentions true where any entries are missing
df.isnull()
# I kept 4 entries of unit price null in csv manually to check
df.isnull().any(axis=1)
df.isnull().sum() # In which column how man entries are missing
```

```python
# Fill empty entries with something
df.fillna(0)
```

```python
## Filling missing values with the mean of the values
df['Unit_Price_fillNA'] = df['Unit Price'].fillna(df['Unit Price'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming Column name  (.rename())
# Syntax data.rename(columns={'PrevName': 'NewName'})

df = df.rename(columns={'Country': 'Nationality'})
df
```

```python
df.dtypes
```

```python
## Changing datatypes (.astype())
# lets change units sold from int to float
df['New Units Sold'] = df['Units Sold'].astype(float)
df.head()
```

```python
## Applying function on any data  (.apply())
df['New Value']=df['New Units Sold'].apply(lambda x:x*2)
df.head()
```
