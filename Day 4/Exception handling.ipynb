```python
## Exception Try, Except Block
try:
    a = b
except:
    print("The variable 'b' is not defined.")
```

```python
# Actually we can catch the kind of exception (NameError in this case)
a = b
```

```python
# So we can call this exception in except block  (using alais here using as keyword)
try:
    a=b
except NameError as ex:
    print(ex)
```

```python
try:
    result = 1/0
except ZeroDivisionError as err:
    print("Enter denom greater than 0")
```

```python
# Here only ZeroDivisionError will be caught not the NameError cuz i have only caught that in except block
try:
    result = 1/2
    a = b
except ZeroDivisionError as err:
    print("Enter denom greater than 0")
```

```python
# All types of errors are derivered from single "Exception" named class
# if we use "Exception" class we have to use it in the end after all other error classes/types
try:
    result = 1/2
    a = b
except ZeroDivisionError as err:
    print("Enter denom greater than 0")
except Exception as nmer:
    print(nmer)
```

```python
# Example of try except
try:
    numb = int(input("Enter a value: "))
    res = 10/numb
except ValueError:
    print("Enter valid value")
except ZeroDivisionError:
    print("Enter greater than 0")
except Exception as ex:
    print(ex)
```

```python
# try, except and else block
# else will be executed only and only if try executes successfully without triggering any except block. 
# if any except block is triggered then that except will be excuted and else will not be
try:
    numb = int(input("Enter a value: "))
    res = 10/numb
except ValueError:
    print("Enter valid value")
except ZeroDivisionError:
    print("Enter greater than 0")
except Exception as ex:
    print(ex)
else:
    print(f"the value is {res}")
```

```python
# try, except, else and finally block
# So finally block is exectued no mattter what happens. even if try block is successfull and else runs or any of the except block is triggered still finally will be executed
# So program runs in try block if successfull then else block after that finally block is executed. If there is exception then that except block is executed and then finally is executed

try:
    numb = int(input("Enter a value: "))
    res = 10/numb
except ValueError:
    print("Enter valid value")
except ZeroDivisionError:
    print("Enter greater than 0")
except Exception as ex:
    print(ex)
else:
    print(f"the value is {res}")
finally:
    print("Execution completed")
```
