  - Examples of process and thread

- Program -> A program is a sequence of instructions written in programming language.
    - Eg: Google Chrome -> .exe <- Sequence of instructions
- Process -> A process is simply an instance of a program that is bieng executed.
    - A process has Code Segment, Data Segment, Heap, Stack, Registers.
    - Eg: Opening Google Chrome
    - Every process will allocate seperate memory space
            - One process cannot corrupt another process
    - More processes will require more I/O resources -> Leading to increased time to switch between processes
- Threads -> A thread is a unit of execution within the process
    - Types: 
    - Single threaded process -> These share the code segment, data segment and heap from the process but it will have its own stack and registers. It is like a sub process(from my understanding)
    - Multi threaded process -> Every thread in a process will have its own stack and register within the process but share code segment, data segment and heap within the process


## Multithreading in Python

#### When To use Multithreading
- I/O-Bound Tasks: Tasks that spend more time waiting for I/O operations (eg: File operations, Network requests).
- Concurrent Execution: When you want to improve the throughput of your application by performing multiple operations concurrently

```python
# Example:
import threading
import time

def print_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"Number: {i}")

def print_letters():
    for letter in "abcde":
        time.sleep(2)
        print(f"Letter: {letter}")

# Before Multithreading
start_time = time.time()
print_letters()
print_numbers()
finished_time = time.time() - start_time
print(f"Finished in {finished_time:.4f} seconds")

```

#### Here, we first run two functions sequentially on a single thread.

Each function includes `sleep(2)`, so while one function is sleeping, the program does not execute the other function. This is why the total runtime (`finished_time`) is roughly the sum of both functions.

With **multithreading**, we can run these functions in separate threads within the same process. Then, when one thread is waiting (sleeping or doing I/O), the other thread can continue executing. This improves concurrency for I/O-bound tasks and usually reduces overall execution time compared to single-threaded execution.

```python
# Example with Multithreading:
import threading
import time

def print_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"Number: {i}")

def print_letters():
    for letter in "abcde":
        time.sleep(2)
        print(f"Letter: {letter}")

## Creating 2 threads
t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

# After Multithreading
start_time = time.time()
t1.start()
t2.start()

# Wait for the threads to complete
t1.join()
t2.join()

finished_time = time.time() - start_time
print(f"Finished in {finished_time:.4f} seconds")

```

So we save 10 seconds by using multithreading
